```python
##new version for liver tests------------------------------------------------------------------------
##importing packages---------------------------------------------------------------------------------
import json
import pandas as pd
import math
from termcolor import colored

##---------------------------------------------------------------------------------------------------
##loading files--------------------------------------------------------------------------------------
with open('factors.json') as json_file:
    Factors = json.load(json_file)
input_df = pd.read_excel("input_liver.xlsx")

##---------------------------------------------------------------------------------------------------
##finding abnormalities------------------------------------------------------------------------------
low = pd.DataFrame()
high = pd.DataFrame()
for i in range(2,len(input_df)):
  if input_df.at[i, 'amount'] < input_df.at[i,'Lower Limit']:
    low = low.append(input_df.iloc[i])
  elif input_df.at [i, 'amount'] > input_df.at[i,'Upper Limit']:
    high = high.append(input_df.iloc[i])   
low = low.reset_index(drop=True)
high = high.reset_index(drop=True)

##---------------------------------------------------------------------------------------------------
##finding factors that we can interpert (high)-------------------------------------------------------
DoNotHave = []
Find = False
for i in range(len(high)):
  for factor in Factors:
    if high.at[i,'parameters'].upper() in Factors[factor]['name']:
      if Factors[factor]['abnormality']['high'] != None:
        high.at[i,'parameters'] = factor + 'High'
        Find = True
  if Find == False:
    DoNotHave.append('high '+ high.at[i,'parameters'])
    high = high.drop(i)
  else:
    Find = False

high = high.reset_index(drop=True)
##finding factors that we can interpert (low)--------------------------------------------------------
Find = False
for i in range(len(low)):
  for factor in Factors:
    if low.at[i,'parameters'].upper() in Factors[factor]['name']:
      if Factors[factor]['abnormality']['low'] != None:
        low.at[i,'parameters'] = factor + 'Low'
        Find = True
  if Find == False:
    DoNotHave.append('low ' + low.at[i,'parameters'])
    low = low.drop(i)
  else:
    Find = False
  
low = low.reset_index(drop=True)

##printing factors that we can not interper----------------------------------------------------------
if len(DoNotHave) > 0 :
  print('The application can not interpert the following factors:')
  for i in range(len(DoNotHave)):
    print('\u2022 ' + DoNotHave[i])

##---------------------------------------------------------------------------------------------------
##creating objects-----------------------------------------------------------------------------------
class ResultFactors:
  def __init__(self, Name, MultipleFactors, QuestionBased, ValueBased, Operation, Obtain, Question, Computation, Suggestion):
    #self.ReferenceRange = ReferenceRange
    self.Name = Name  
    self.MultipleFactors = MultipleFactors
    self.QuestionBased = QuestionBased
    self.ValueBased = ValueBased
    self.Operation = Operation
    self.Obtain = Obtain
    self.Question = Question
    self.Computation = Computation
    self.Suggestion = Suggestion

##creat object for high------------------------------------------------------------------------------
FactorObjectHigh ={}    
Name = []
MultipleFactors = []
QuestionBased = []
ValueBased = []
Operation =[]
Obtain = []
Question = []
Computation = []
Suggestion = []

for factor in Factors:
  MultipleFactors.append(Factors[factor]['multiple factors'])
  if Factors[factor]['abnormality']['high'] != None:
    for i in range(len(Factors[factor]['abnormality']['high'])):
      Name.append(Factors[factor]['abnormality']['high'][i]['Name'])
      QuestionBased.append(Factors[factor]['abnormality']['high'][i]['Question based'])
      ValueBased.append(Factors[factor]['abnormality']['high'][i]['Value based'])
      Operation.append(Factors[factor]['abnormality']['high'][i]['Operation'])
      Obtain.append(Factors[factor]['abnormality']['high'][i]['Obtain'])
      Question.append(Factors[factor]['abnormality']['high'][i]['Question'])
      Computation.append(Factors[factor]['abnormality']['high'][i]['Computation'])
      Suggestion.append(Factors[factor]['abnormality']['high'][i]['Suggestion'])
    
    ObjectName = factor + 'High'
    FactorObjectHigh[ObjectName] = FactorObjectHigh.get(ObjectName, ResultFactors(Name, MultipleFactors, QuestionBased, ValueBased, Operation, Obtain, Question, Computation, Suggestion))
    Name = []
    MultipleFactors = []
    QuestionBased = []
    ValueBased = []
    Operation =[]
    Obtain = []
    Question = []
    Computation = []
    Suggestion = []

##creat object for low------------------------------------------------------------------------------
FactorObjectLow ={}    
Name = []
MultipleFactors = []
QuestionBased = []
ValueBased = []
Operation =[]
Obtain = []
Question = []
Computation = []
Suggestion = []

for factor in Factors:
  MultipleFactors.append(Factors[factor]['multiple factors'])
  if Factors[factor]['abnormality']['low'] != None:
    for i in range(len(Factors[factor]['abnormality']['low'])):
      Name.append(Factors[factor]['abnormality']['low'][i]['Name'])
      QuestionBased.append(Factors[factor]['abnormality']['low'][i]['Question based'])
      ValueBased.append(Factors[factor]['abnormality']['low'][i]['Value based'])
      Operation.append(Factors[factor]['abnormality']['low'][i]['Operation'])
      Obtain.append(Factors[factor]['abnormality']['low'][i]['Obtain'])
      Question.append(Factors[factor]['abnormality']['low'][i]['Question'])
      Computation.append(Factors[factor]['abnormality']['low'][i]['Computation'])
      Suggestion.append(Factors[factor]['abnormality']['low'][i]['Suggestion'])
    
    
    ObjectName = factor + 'Low'
    FactorObjectLow[ObjectName] = FactorObjectLow.get(ObjectName, ResultFactors(Name, MultipleFactors, QuestionBased, ValueBased, Operation, Obtain, Question, Computation, Suggestion))
    Name = []
    MultipleFactors = []
    QuestionBased = []
    ValueBased = []
    Operation =[]
    Obtain = []
    Question = []
    Computation = []
    Suggestion = []

##---------------------------------------------------------------------------------------------------
##Functions------------------------------------------------------------------------------------------
##opperation does simple math operation based on its inputs------------------------------------------
def operation(Num1, Num2, operator):
  if operator == 'multiplication':
    return Num1*Num2
  elif operator == 'division':
    return Num1/Num2
  elif operator == 'sum':
    return Num1+Num2
  elif operator == 'subtraction':
    return Num1-Num2
  elif operator == 'less':
    if Num1 < Num2:
      return True
    else:
      return False
  elif operator == 'greater':
    if Num1 > Num2:
      return True
    else:
      return False
##end of function------------------------------------------------------------------------------------
#compelet later



##---------------------------------------------------------------------------------------------------
#compelet later
if len(low) > 0:
  for i in range(len(low)):
    print(FactorObjectLow[low.at[i, 'parameters']].Name)

if len(high) > 0:
  for i in range(len(high)):
    print(FactorObjectHigh[high.at[i, 'parameters']].Name)
```
